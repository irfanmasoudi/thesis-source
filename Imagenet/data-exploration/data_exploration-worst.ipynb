```python
import numpy as np
import pandas as pd
```

```python
df_original = pd.read_excel("cm_original_1.xlsx", sheet_name='Sheet_name_1') 
df_original.head()
```

```python
columns =['Label', 'True', 'False', 'Accuracy']
lst = []
for i, row in df_original.iterrows():
    label_n = df_original.iloc[i,0]
    true_number = df_original.iloc[i,i+1]
    error_number = 25 - true_number
    accuracy = (true_number/25) * 100
    lst.append([label_n, true_number, error_number, accuracy])

df_ori_exp = pd.DataFrame(lst, columns=columns)
df_ori_exp.head()
```

```python
df_ori_worst10 = df_ori_exp.sort_values(by=['Accuracy'], ascending=True).iloc[:10,:]
df_ori_worst10_save = df_ori_worst10.reset_index(drop=True)
df_ori_worst10_save.index = np.arange(1, len(df_ori_worst10)+1)
df_ori_worst10_save.to_csv('worst10_ori.csv')
df_ori_worst10
```

```python
df_ori_worst20 = df_ori_exp.sort_values(by=['Accuracy'], ascending=True).iloc[:20,:]
df_ori_worst20_save = df_ori_worst20.reset_index(drop=True)
df_ori_worst20_save.index = np.arange(1, len(df_ori_worst20)+1)
df_ori_worst20_save.to_csv('worst20_ori.csv')
df_ori_worst20
```

```python
df_ori_worst50 = df_ori_exp.sort_values(by=['Accuracy'], ascending=True).iloc[:50,:]
df_ori_worst50_save = df_ori_worst50.reset_index(drop=True)
df_ori_worst50_save.index = np.arange(1, len(df_ori_worst50)+1)
df_ori_worst50_save.to_csv('worst50_ori.csv')
df_ori_worst50
```

```python
df_ori_worst100 = df_ori_exp.sort_values(by=['Accuracy'], ascending=True).iloc[:100,:]
df_ori_worst100_save = df_ori_worst100.reset_index(drop=True)
df_ori_worst100_save.index = np.arange(1, len(df_ori_worst100)+1)
df_ori_worst100_save.to_csv('worst100_ori.csv')
df_ori_worst100
```
