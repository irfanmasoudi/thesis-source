```python
import numpy as np
import pandas as pd
```

```python
df_original = pd.read_excel("cm_original_1.xlsx", sheet_name='Sheet_name_1') 
df_original.head()
```

```python
columns =['Label', 'True', 'False', 'Accuracy']
lst = []
for i, row in df_original.iterrows():
    label_n = df_original.iloc[i,0]
    true_number = df_original.iloc[i,i+1]
    error_number = 25 - true_number
    accuracy = (true_number/25) * 100
    lst.append([label_n, true_number, error_number, accuracy])

df_ori_exp = pd.DataFrame(lst, columns=columns)
df_ori_exp.head()
```

```python
df_ori_top10 = df_ori_exp.sort_values(by=['Accuracy'], ascending=False).iloc[:10,:]
df_ori_top10_save = df_ori_top10.reset_index(drop=True)
df_ori_top10_save.index = np.arange(1, len(df_ori_top10)+1)
df_ori_top10_save.to_csv('top10_ori.csv')
df_ori_top10
```

```python
df_ori_top20 = df_ori_exp.sort_values(by=['Accuracy'], ascending=False).iloc[:20,:]
df_ori_top20_save = df_ori_top20.reset_index(drop=True)
df_ori_top20_save.index = np.arange(1, len(df_ori_top20)+1)
df_ori_top20_save.to_csv('top20_ori.csv')
df_ori_top20
```

```python
df_ori_top50 = df_ori_exp.sort_values(by=['Accuracy'], ascending=False).iloc[:50,:]
df_ori_top50_save = df_ori_top50.reset_index(drop=True)
df_ori_top50_save.index = np.arange(1, len(df_ori_top50)+1)
df_ori_top50_save.to_csv('top50_ori.csv')
df_ori_top50.head()
```

```python
df_ori_top100 = df_ori_exp.sort_values(by=['Accuracy'], ascending=False).iloc[:100,:]
df_ori_top100_save = df_ori_top100.reset_index(drop=True)
df_ori_top100_save.index = np.arange(1, len(df_ori_top100)+1)
df_ori_top100_save.to_csv('top100_ori.csv')
df_ori_top100
```

```python
import re 
from tqdm.notebook import tqdm

columns =['label','name', 'augmented true', 'augmented false']
lst_true = []
lst_false = []

pattern = r'n\d{8}_\d+'
df_regex = pd.DataFrame()

for i, row in tqdm(df_list_test_ori.iterrows()):
    label = row['label']
    value = row['T/F']
    file_name_ori = row['name']
    match = re.search(pattern, file_name_ori)
    result = match.group(0)
    result = f"{result}_"
    df_regex = df_list_test_aug.filter(like=result, axis=0)
    true = df_regex['T/F'].values.sum()
    # if true > 5:
    #     true = 5
    false = 5 - true
    # print(i)
    if value == True:
        lst_true.append([label, file_name_ori, true, false])
    # break
    else:
        lst_false.append([label, file_name_ori, true, false])
        
df_true_result = pd.DataFrame(lst_true, columns=columns)
df_false_result = pd.DataFrame(lst_false, columns=columns)
df_true_result.to_excel("images_ori_pred_true_2.xlsx", sheet_name='Sheet_name_1') 
df_false_result.to_excel("images_ori_pred_false_2.xlsx", sheet_name='Sheet_name_1') 
```
