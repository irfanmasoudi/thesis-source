```python
import pandas as pd

df_cm = pd.read_excel("confmat_2.xlsx", index_col=0) 
df_cm
```

```python
total_samples = df_cm.values.sum()
confusion_matrix_percentage = (df_cm / total_samples * 100).round(2)

confusion_matrix_percentage
```

```python
labels = pd.read_csv("label-shift.csv").iloc[:,0].tolist()
labels
```

```python
confusion_matrix_percentage = confusion_matrix_percentage.rename(columns=dict(zip(confusion_matrix_percentage.columns, labels)),
                                                 index=dict(zip(confusion_matrix_percentage.index, labels)))
```

```python
confusion_matrix_percentage.to_excel('cm_percentage.xlsx')
```

```python
cm_percent_row = (df_cm.div(df_cm.sum(axis=1), axis=0) * 100).round(2)
cm_percent_row
```

```python
cm_percent_row = cm_percent_row.rename(columns=dict(zip(cm_percent_row.columns, labels)),
                                                    index=dict(zip(cm_percent_row.index, labels)))
```

```python
cm_percent_row.to_excel('cm_percent_row.xlsx')
```
